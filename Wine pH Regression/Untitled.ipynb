$F=\frac{2 \cdot \text { precision } \cdot \text {recall}}{\text {precision}+\text {recall}}$

```python
import pylab as plt
import numpy
b=2

def makeData():
    x = numpy.arange(0, 1.05, 0.01) #precision
    y = numpy.arange(0, 1.05, 0.05) #recall
    xgrid, ygrid = numpy.meshgrid(x, y)

    zgrid = ((1+b**2)*xgrid*ygrid)/(b**2*xgrid+ygrid)
    return xgrid, ygrid, zgrid


plt.figure(figsize=(9,9))
x, y, z = makeData()
cs = plt.contour(x, y, z)
plt.clabel(cs)


plt.xlabel(u'Точность', fontsize=22)
plt.ylabel(u'Полнота', fontsize=22)
plt.xticks(fontsize=22)
plt.yticks(fontsize=22)

plt.xlim(0.01,1.00001)
plt.ylim(ymax=1.0001)
plt.grid()
#plt.show()
plt.savefig('f1_score_example_b_2.pdf')
```
